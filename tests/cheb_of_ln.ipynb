```python
from pychebfun import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.arange(-1.00,0.5,0.01)
def ll(x):
    return np.log(0.02+1+(x-3)/4)
g = Chebfun.from_function(ll,N=17)
plt.plot(np.log(np.abs(g.coefficients()))/np.log(10))
plt.show()
print(g.coefficients())
print(len(g.coefficients()))
# plot(g)
plt.plot(xx,ll(xx)-g(xx))
# plt.plot(xx,g(xx))
plt.show()
```
